```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import itertools
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%matplotlib inline
```

```python
booking=pd.read_excel('bookings.csv',index_col=0)
```

```python
look_up = {'Jan':'01', 'Feb':'02' , 'Mar':'03', 'Apr':'04', 'May':'05','Jun':'06', 'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
booking['Month'] = booking['Month'].apply(lambda x: look_up[x])
```

```python
booking["date"] = booking.index.map(str) + "-"+booking["Month"]
booking.date = pd.to_datetime(booking.date)
booking = booking.set_index("date")
booking.drop("Month", axis=1, inplace=True)
```

```python
booking.head(5)
```

```python
booking.plot()
```

```python
decomposition = seasonal_decompose(booking,freq=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
booking.plot(figsize=(16, 3))
trend.plot(figsize=(16, 3))
seasonal.plot(figsize=(16, 3))
residual.plot(figsize=(16, 3))
```

```python
(trend+seasonal+residual).plot()
```

# make the TS stationary season-trend

```python
season_diff = booking.diff(12).dropna()
season_diff1 = season_diff.diff(1).dropna()
season_diff1.plot()
```

```python
plot_acf(season_diff1,lags=24);
```

```python
plot_pacf(season_diff1,lags=36);
```

The acf and pacf plot indicate both, seasonality and trend components.

## now the sarimax

```python
model = SARIMAX(booking, order=(0,1,1), seasonal_order=(0,1,1,12),enforce_stationarity=False,enforce_invertibility=False)
results = model.fit()
```

```python
print(results.aic)
```

```python
pred = results.get_prediction(start=pd.to_datetime('2015-01-01'), dynamic=False)
fore = results.get_forecast(12)
```

```python
booking.plot()
pred.predicted_mean.plot()
fore.predicted_mean.plot()
```

# Make the TS stationary from independent analysis

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
AIC=[]
params=[]
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            model = SARIMAX(booking,order=param,seasonal_order=param_seasonal,enforce_stationarity=False,enforce_invertibility=False)
            results = model.fit()
            params.append(str(param)+" x "+str(param_seasonal))
            AIC.append(results.aic)
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
params[np.argmin(AIC)], np.min(AIC)
```

```python
model = SARIMAX(booking, order=(0,1,1), seasonal_order=(0,1,1,12),enforce_stationarity=False,enforce_invertibility=False)
results = model.fit()
```

```python
pred = results.get_prediction(start=pd.to_datetime('2015-01-01'), dynamic=False)
fore = results.get_forecast(36)
fore_ci = fore.conf_int()
```

```python
ax = booking.plot()
pred.predicted_mean.plot(ax=ax)
fore.predicted_mean.plot(ax=ax)
ax.fill_between(fore_ci.index,
                fore_ci.iloc[:, 0],
                fore_ci.iloc[:, 1], color='k', alpha=.25)
```
