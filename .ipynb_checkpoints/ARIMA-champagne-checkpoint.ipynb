```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import itertools
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%matplotlib inline
```

```python
sales=pd.read_excel('champagne-sales.xlsx',index_col=0)
```

```python
sales = sales.reset_index()
sales["Month"] = "200"+sales.Month.map(str)
sales.Month = pd.to_datetime(sales.Month)
sales = sales.set_index("Month")
```

```python
sales.head(5)
```

```python
decomposition = seasonal_decompose(sales,freq=12)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
sales.plot(figsize=(16, 3))
trend.plot(figsize=(16, 3))
seasonal.plot(figsize=(16, 3))
residual.plot(figsize=(16, 3))
```

```python
season_diff = sales.diff(12).dropna()
season_diff1 = season_diff.diff(1).dropna()
season_diff.plot()
```

```python
sales.tail(6)
```

```python
plot_acf(season_diff1,lags=24);
```

```python
plot_pacf(season_diff1,lags=24);
```

p=1, d=0, q=0, P=1, D=1, Q=0, m=12

```python
model = SARIMAX(sales, order=(1,0,0), seasonal_order=(1,1,0,12),enforce_stationarity=False,enforce_invertibility=False)
results = model.fit()
print(results.aic)
```

```python
pred = results.get_prediction(start=pd.to_datetime('2008-07-01'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
ax=sales[-6:].plot()
pred.predicted_mean.plot()
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
```

```python
pred.predicted_mean
```

```python
sales_train = sales[:-6]
```

```python
model = SARIMAX(sales_train, order=(1,0,0), seasonal_order=(1,1,0,12),enforce_stationarity=False,enforce_invertibility=False)
results = model.fit()
print(results.aic)
```

```python
fore = results.get_forecast(6)
fore_ci = fore.conf_int()
```

```python
ax=sales[-6:].plot()
fore.predicted_mean.plot()
ax.fill_between(fore_ci.index,
                fore_ci.iloc[:, 0],
                fore_ci.iloc[:, 1], color='k', alpha=.25)
```
