```python
import rpy2.robjects as robjects
from rpy2.robjects.packages import importr
ts=robjects.r('ts')
forecast=importr('forecast')
import pandas as pd
from rpy2.robjects import pandas2ri
pandas2ri.activate()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales=pd.read_excel('champagne-sales.xlsx',index_col=0)
```

```python
sales = sales.reset_index()
sales["Month"] = "200"+sales.Month.map(str)
sales.Month = pd.to_datetime(sales.Month)
sales = sales.set_index("Month")
```

```python
sales.tail(6)
```

```python
sales.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(sales["Champagne Sales"].to_frame(),freq=12)
```

```python
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
sales["Champagne Sales"].plot(figsize=(16, 3))
trend.plot(figsize=(16, 3))
seasonal.plot(figsize=(16, 3))
residual.plot(figsize=(16, 3))
```

```python
rdata=ts(sales["Champagne Sales"].head(90).values,frequency=1)
```

```python
rstring="""
    function(testdata){
        library(forecast)
        fitted_model<-ets(testdata,model="MAM")
        forecasted_data<-forecast(fitted_model,h=6,level=c(95))
        outdf<-data.frame(forecasted_data$mean,forecasted_data$lower,forecasted_data$upper)
        colnames(outdf)<-c('forecast','lower_95_pi','upper_95_pi')
        outdf
    }
"""
rfunc=robjects.r(rstring)

rdata=ts(sales["Champagne Sales"].values,frequency=12)
r_df=rfunc(rdata)
```

```python
forecast_df=pandas2ri.ri2py(r_df)
forecast_df.index=pd.date_range('2008-12-01',periods=len(forecast_df)+1,freq='M')[:-1]

forecast_df
```

```python
fig=plt.figure(figsize=(16, 7));
ax=plt.axes()
ax.plot(sales.index,sales["Champagne Sales"].values,color='blue',alpha=0.5)
ax.plot(forecast_df.index,forecast_df.forecast.values,color='red')
ax.fill_between(forecast_df.index,
                      forecast_df['lower_95_pi'],
                      forecast_df['upper_95_pi'],
                      alpha=0.2,color='red')
```
